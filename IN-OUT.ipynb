```python
# apenas um linha por cod referencia
df_unique = df_result.drop_duplicates(subset='REFERENCIA').copy()


def convert_to_float(value):
    try:
        return float(value)
    except ValueError:
        return None


# aplicar as restrições
def verificar_restricoes(codes_all, esp, comp, larg, restricoes):
    valid_codes = []
    esp = convert_to_float(esp)
    comp = convert_to_float(comp)
    larg = convert_to_float(larg)

    for code in codes_all:
        restricao = restricoes[restricoes['Código '].str.strip() == code]
        if restricao.empty:
            continue

        min_esp = convert_to_float(restricao['Espessura mínima'].values[0])
        max_esp = convert_to_float(restricao['Espessura máxima'].values[0])
        min_dim = convert_to_float(restricao['Dimensão mínima'].values[0])
        max_dim = convert_to_float(restricao['Dimensão máxima'].values[0])

        # restrições
        if min_esp is not None and max_esp is not None and not (min_esp <= esp <= max_esp):
            continue

        maior_dim = max(comp, larg)
        if max_dim is not None and not (maior_dim <= max_dim):
            continue

        menor_dim = min(comp, larg)
        if min_dim is not None and not (menor_dim >= min_dim):
            continue

        valid_codes.append(code)

    return valid_codes

#função para cada linha no DataFrame, usando .loc
df_unique.loc[:, 'CODES_all'] = df_unique.apply(
    lambda row: verificar_restricoes(
        row['CODES_all'], row['ESP'], row['COMP'], row['LARG'], df_restricoes
    ),
    axis=1
)
```
